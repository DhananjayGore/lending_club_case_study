```python
#Bucketing different varaibles for better analysis
# Bucketting Loan Amount
def bucket_loan_amnt(column):
    if column <= 5000:
        return '0 - 5K'  # 25% quartile
    elif (column >5000) and (column <= 10000):
        return '5K - 10K'
    elif (column >10000) and (column <= 15000):
        return '10K - 15K'
    else:
        return '15K - above' # 75% quartile

Lending_Club_Loan_Cleaned['loan_amnt_b'] = Lending_Club_Loan_Cleaned.apply(lambda x : bucket_loan_amnt(x['loan_amnt']), axis = 1)
```

```python
#Bucketing Annual Income
def bucket_annual_inc(column):
    if column <= 40000:
        return '0 - 40k' # 25% quartile
    elif (column >40000) and (column <= 50000):
        return '40k - 50k'
    elif (column >50000) and (column <= 60000):
        return '50k to 60k'
    elif (column >60000) and (column <= 70000):
        return '60k to 70k'
    elif (column >70000) and (column <= 80000):
        return '70k to 80k'
    else:
        return '80k - above' # 75% quartile

Lending_Club_Loan_Cleaned['annual_inc_b'] = Lending_Club_Loan_Cleaned.apply(lambda x: bucket_annual_inc(x['annual_inc']), axis = 1)
```

```python
# Bucketing interest rate
def bucket_int_rate(column):
    if column <= 9:
        return 'Very Low' # 25% quartile
    elif (column >9) and (column <= 11):
        return 'Low'
    elif (column >11) and (column <= 13):
        return 'Moderate'
    elif (column >13) and (column <= 15):
        return 'High'
    else:
        return 'Very High' # 75% quartile

Lending_Club_Loan_Cleaned['int_rate_b'] = Lending_Club_Loan_Cleaned.apply(lambda x : bucket_int_rate(x.int_rate), axis = 1)
```

```python
# Bucketing dti
def bucket_dti(column):
    if column <= 8:
        return 'Very Low' # 25% quartile
    elif (column >8) and (column <= 12):
        return 'Low'
    elif (column >12) and (column <= 16):
        return 'Moderate'
    elif (column >16) and (column <= 20):
        return 'High'
    else:
        return 'Very High' # 75% quartile

Lending_Club_Loan_Cleaned['dti_b'] = Lending_Club_Loan_Cleaned.apply(lambda x : bucket_dti(x.dti), axis = 1)
```

```python
# Univarients function this function will help in plotting the chart
# Updated function to plot diagrams using histplot and boxplot
def plot_diagram(df, column, column_titles):
    fig, p = plt.subplots(1, 2, figsize=(16, 4))
    
    # Plot histogram with KDE
    sns.histplot(df[column].dropna(), kde=True, ax=p[0])
    sns.boxplot(x=column, data=df, ax=p[1])
    
    p[0].set_xlabel(column_titles.get(column, column))
    p[1].set_xlabel(column_titles.get(column, column))
    
    plt.tight_layout()
    plt.show()
```

```python
# From above box plts we can see there is otliers exits for columns'loan_amnt', 'funded_amnt', 'funded_amnt_inv', 'int_rate', 'installment', 'annual_inc'.
# Lets remove the outlier
#Interquartile Range (IQR)
outliers_columns_set1= ['loan_amnt', 'funded_amnt', 'funded_amnt_inv']
outliers_columns_set2=['int_rate', 'installment', 'annual_inc']
# Define a function to remove outliers based on IQR
def remove_outliers_iqr(data, columns,threshold):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data

# Remove outliers from the DataFrame
loan_df =remove_outliers_iqr(Lending_Club_Loan_Cleaned, outliers_columns_set1,1.5)
loan_df =remove_outliers_iqr(Lending_Club_Loan_Cleaned, outliers_columns_set2,1.5)
```

### <font color = "blue"> Analysis on Charged Off, Current and Fully Paid loan distribution across grades </font >

```python
LC_status_grade=Lending_Club_Loan_Cleaned.groupby('loan_status')['grade'].value_counts().unstack(fill_value=0)
LC_status_grade_charged_off = LC_status_grade.loc['Charged Off']
LC_status_grade_Current = LC_status_grade.loc['Current']
LC_status_grade_FullyPaid = LC_status_grade.loc['Fully Paid']
LC_Total_LoanByGrade=Lending_Club_Loan_Cleaned['grade'].value_counts()
#LC_PercentageByGrade = (LC_status_grade_charged_off / LC_Total_LoanByGrade.sum()) * 100
fig, axes = plot.subplots(nrows=5, ncols=1, figsize=(24, 32))
charged_off_percentages = {}

for grade in LC_status_grade_charged_off.index:
    charged_off_count = LC_status_grade_charged_off[grade]
    total_count = LC_Total_LoanByGrade[grade]   
    percentage = (charged_off_count / total_count) * 100 
    charged_off_percentages[grade] = percentage
grades = list(charged_off_percentages.keys())
percentages = list(charged_off_percentages.values())


axes[0].bar(LC_Total_LoanByGrade.index, LC_Total_LoanByGrade.values, color='skyblue', edgecolor='black')
axes[0].set_title('Loans by Grade')
axes[0].set_ylabel('Number of Loans Sanctioned')
axes[0].set_xticks(LC_Total_LoanByGrade.index) 
axes[0].set_xlabel('Grade')
for i in range(len(LC_Total_LoanByGrade)):
    axes[0].text(LC_Total_LoanByGrade.index[i], LC_Total_LoanByGrade.values[i] + 0.1, str(LC_Total_LoanByGrade.values[i]), ha='center')

axes[1].bar(LC_status_grade_FullyPaid.index, LC_status_grade_FullyPaid.values, color='green', edgecolor='black')
axes[1].set_title('Fully Paid Loans By Grade')
axes[1].set_ylabel('Number of Loans Sanctioned')
axes[1].set_xlabel('Grade')
for i in range(len(LC_status_grade_FullyPaid)):
    axes[1].text(LC_status_grade_FullyPaid.index[i], LC_status_grade_FullyPaid.values[i] + 0.1, str(LC_status_grade_FullyPaid.values[i]), ha='center')

axes[2].bar(LC_status_grade_Current.index, LC_status_grade_Current.values, color='red', edgecolor='black')
axes[2].set_title('Current Ongoing Loans by Grade')
axes[2].set_ylabel('Number of Loans Sanctioned')
axes[2].set_xlabel('Grade')
for i in range(len(LC_status_grade_Current)):
    axes[2].text(LC_status_grade_Current.index[i], LC_status_grade_Current.values[i] + 0.1, str(LC_status_grade_Current.values[i]), ha='center')

axes[3].bar(LC_status_grade_charged_off.index, LC_status_grade_charged_off.values, color='red', edgecolor='black')
axes[3].set_title('Charged off loans By Grade')
axes[3].set_ylabel('Number of Loans Sanctioned')
axes[3].set_xlabel('Grade')
for i in range(len(LC_status_grade_charged_off)):
    axes[3].text(LC_status_grade_charged_off.index[i], LC_status_grade_charged_off.values[i] + 0.1, str(LC_status_grade_charged_off.values[i]), ha='center')

axes[4].bar(grades, percentages, color='pink', edgecolor='black')
axes[4].set_title('Charged off loans % By Grade')
axes[4].set_ylabel('%loans Charged off by  grade')
axes[4].set_xlabel('Grade')
for grade, percentage in zip(grades, percentages):
    axes[4].text(grade, percentage + 0.1, f'{percentage:.1f}%', ha='center', va='bottom', fontsize=10)
```
